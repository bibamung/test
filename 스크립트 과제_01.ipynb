### 5. for문을 활용하여 1부터 100사이의 홀수를 출력하시오.

```python
for i in range(1,100):
    if i % 2 != 0:
        print(i, end=" ")
```

for문을 사용하여 i값의 범위를 설정하고 i를 2로 나눴을때 나눠떨어지지 않으면 출력함

### 6. while문을 활용하여 1부터 100사이의 짝수의 합을 계산하여 출력하시오.

```python
n=1
while(n <= 100):
    if(n%2==0):
        print(n, end = " ")
    n+=1
```

while문을 사용하여 n값으로 얼마나 반복시킬 것인지 설정하고 n를 2로 나눴을때 나눠떨어지지 않으면 출력함

```python
NUM2 = int(input("정수를 입력하세요 >> "))
for i in range (2,NUM2):
    result = True
    if i < 2:
        result = False
    for j in range(2,i):
        if i % j == 0:
            result = False
    if result:
        print(i, end = " ")

    
```

### 9. Kevin과 Stuart는 게임을 한다. 규칙은 아래와 같다.
* 두 플레이어에게 같은 string, S, 가 주어진다.
* 두 플레이어는 string S의 문자를 사용하여 substring을 만들어야 한다(중복 가능).
* Stuart는 자음으로 시작하는 단어를 만들어야 한다.
* Kevin은 모음으로 시작하는 단어를 만들어야 한다.
* 게임은 두 플레이어가 만들 수 있는 모든 단어를 만들었을 때 끝난다.
* 점수 계산 : substring 1개당 +1 점을 획득한다. (아래 그림 참고)
* 
* 사용자로 부터 대문자로 이루어진 string S를 입력받고, 승자와 그의 점수를 출력하는 프로그램을 작성하시오.
####    [실행 예]    
        string S를 입력하세요: BANANA 
        승자: Stuart 점수: 12          

```python
S = input("string S를 입력하세요 : ")
S_Stuart = ["Stuart", 0]
S_Kevin = ["Kevin", 0]
for i in range(len(S)):    #문자열 S만큼
    if S[i]=='A' or S[i]=='E' or S[i]=='O' or S[i]=='I' or S[i]=='U':
        S_Kevin[1] += len(S)-i
    else:
        S_Stuart[1] += len(S)-i
if S_Stuart[1] < S_Kevin[1]:
    print("승자 : ",S_Kevin[0], "점수 :", S_Kevin[1])
elif S_Stuart[1] > S_Kevin[1]:
    print("승자 : ",S_Stuart[0], "점수 :", S_Stuart[1])
else:
    print("무승부입니다...")
```

플레이어의 이름과 플레이어의 점수를 저장할수 있는 list를 생성  
Kevin은 모음으로 시작하는 단어를 만들어야 점수를 얻을 수 있기때문에 조건문을 활용해 S[i]의 위치가 모음인 단어를 생성할때마다 점수를 획득하게 함.  
반대로 자음으로 시작하는 단어를 생성할때마다 Stuart에게 점수를 획득할수 있게 함.

그리고 조건문으로 승자를 검사하여 출력하는 구문을 만듦.

### 10. 알파벳 소문자만으로 이루어진 string, S, 가 주어지고 아래 2가지 연산을 수행할 수 있다.
* 추가 : string S의 끝에 알파벳 소문자 하나를 추가한다.
* 삭제 : string S의 마지막 문자 하나를 삭제한다. empty string에서 삭제 연산을 수행한 결과는 empty string이다.

두 개의 string, s와 t, 그리고 정수 k가 주어질 때, 정확히 k번의 연산을 통해 s를 t로 바꿀 수 있는지 판단하는 함수를 작성하시오.

#### [예시]
    1. s = 'hackerhappy'    t = 'hackerrank'    k = 9    => 'Yes'
    
    2. s = 'aba'            t = 'aba'           k = 7    => 'Yes' 
    
    3. s = 'ashley'         t = 'ash'           k = 2    => 'No'

```python
def appendDelete(s,t,k):
    count = 0                 #정확히 몇번만에 가능한지 저장하는 변수
    dif = len(s)              #문자열 s,t의 다른 부분이 시작되는 위치를 저장하는 변수
    s_list = list(s)
    t_list = list(t)
    if k >= (len(s)+len(t)):     #k가 문자열 두개합친 길이보다 길면 실행
        return "'YES'"   
    if len(s_list)>len(t_list) :
        for i in range(len(t)):
            if s_list[i]!=t_list[i]:
                dif=i
                break
    count = len(s_list)-dif
    del s_list[dif:]
    count += len(t_list)-len(s_list)
    count = abs(count)
    if count%2==0:
        if k >= count and k%2==0:
            return "'YES'"
        else:
            return "'NO'"
    else:
        if k >= count and k%2!=0:
            return "'YES'"
        else:
            return "'NO'"
   
   

s = input("단어를 입력하세요 : ")
t = input("단어를 입력하세요 : ")
k = int(input("몇번 안에 바꾸실건지 입력하세요 : "))
print("s =", s, "   t =", t,"   k = ", k, "    =>",  appendDelete(s,t,k))
```
